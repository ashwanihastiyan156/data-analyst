```python
import pandas as pd
import re

# Load Excel file
df = pd.read_excel("ingredient_column.xlsx")

column_name = 'ingredients'

# Updated function: split by whitespace but preserve \n
def remove_hindi_preserve_newlines(text):
    if pd.isna(text):
        return ""

    text = str(text)  # Converts to string
    
    # Split on whitespace including newlines
    words = re.split(r'(\s+)', text)  # Keeps separators like spaces and newlines
    english_words = [word for word in words if not re.search(r'[\u0900-\u097F]', word)]
    return ''.join(english_words)

# Apply to the column
df[column_name] = df[column_name].apply(remove_hindi_preserve_newlines)

# Save to new Excel
df.to_excel("cleaned_output_preserve_newlines.xlsx", index=False)

print("Hindi words removed, newlines preserved.")
```
