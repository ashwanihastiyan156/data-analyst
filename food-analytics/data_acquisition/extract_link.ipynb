```python
import pandas as pd
import re

# Load Excel file
df = pd.read_excel("output_with_ingredients.xlsx")

# Regex pattern to extract the link
pattern = r"https?://hebbarskitchen\.com/[^\s/]+/"

# Function to extract link from description
def extract_link(text, ingredients):
    if pd.isna(text) or pd.isna(ingredients) or str(ingredients).strip() == "":
        return None
    match = re.search(pattern, str(text))
    return match.group(0) if match else None

# Apply function to extract links (check ingredients too)
df["recipe_link"] = df.apply(
    lambda row: extract_link(row["description"], row["ingredients"]),
    axis=1
)

# Keep only required columns (assuming video_id exists in your Excel)
output_df = df[["video_id", "title", "recipe_link"]]

# Export to new Excel
output_df.to_excel("video_descriptions_with_links.xlsx", index=False)

print(" Extraction complete! Saved as 'video_descriptions_with_links.xlsx'")
```
