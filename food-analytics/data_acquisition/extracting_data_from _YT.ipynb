```python
import pandas as pd
from googleapiclient.discovery import build

def get_video_descriptions(video_ids, api_key):
    # Build YouTube API client
    youtube = build("youtube", "v3", developerKey=api_key)

    video_data = []

    # Batch video IDs in groups of 50 (API limit)
    for i in range(0, len(video_ids), 50):
        batch_ids = video_ids[i:i+50]

        # API request
        request = youtube.videos().list(
            part="snippet",
            id=",".join(batch_ids)
        )
        response = request.execute()

        # Extract title and description
        for item in response.get('items', []):
            video_id = item['id']
            snippet = item['snippet']
            title = snippet.get('title', '')
            description = snippet.get('description', '')
            video_data.append({
                'video_id': video_id,
                'title': title,
                'description': description
            })

    return video_data

def main():
    # Input Excel file and API Key
    input_file = 'hebbars.xlsx'
    output_file = 'video_descriptions.xlsx'
    api_key = 'AIzaS**************bNOCcNWTsY'

    # Load video IDs
    df = pd.read_excel(input_file)
    video_ids = df['video_id'].dropna().astype(str).tolist()

    # Get video titles and descriptions
    video_data = get_video_descriptions(video_ids, api_key)

    # Save to Excel
    result_df = pd.DataFrame(video_data)
    result_df.to_excel(output_file, index=False)
    print(f"Video titles and descriptions saved to '{output_file}'")

if __name__ == "__main__":
    main()
```
