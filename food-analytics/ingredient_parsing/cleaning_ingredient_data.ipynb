```python
import pandas as pd
import re

def remove_leading_serials(text):
    if pd.isna(text):
        return ""
    
    lines = str(text).split('\n')  # Split the multi-line cell content
    cleaned_lines = []

    for line in lines:
        # Remove bullets like "1. ", "a) ", "i) ", "2- ", etc.
        cleaned_line = re.sub(r'^\s*([0-9]+|[a-zA-Z]+)[.)-]\s*', '', line)
        cleaned_lines.append(cleaned_line.strip())

    return '\n'.join(cleaned_lines)  # Preserve original line breaks

# === Step 1: Read file ===
input_file = "output_with_ingredients.xlsx"
df = pd.read_excel(input_file)

# === Step 2: Filter only the needed columns ===
filtered_df = df[["video_id", "ingredients"]].copy()

# === Step 3: Clean the ingredients column line-by-line ===
filtered_df["ingredients"] = filtered_df["ingredients"].apply(remove_leading_serials)

# === Step 4: Save the cleaned file ===
output_file = "ingredient_column.xlsx"
filtered_df.to_excel(output_file, index=False)

print(f" Filtered and cleaned data saved to: {output_file}")
```
